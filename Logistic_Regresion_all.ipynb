```python
#build

def sigmoid(z):
    a = 1/(1+np.exp(-z))
    return a

sigmoid(np.array([0,2]))
```

```python
# initializing parameters

def initialize_zeros(dim):
    w = np.zeros((dim,1))
    b = 0
    
    return w,b

def propagate(w,b,X,Y):
    
    #foward
    z = np.dot(w.T, X)+b
    a = sigmoid(z)

    m = X.shape[1]
    loss = (np.dot(Y,np.log(a).T) + np.dot((1-Y),np.log(1-a).T))/-m
    loss = np.squeeze(loss)
    
    #ready for back propa
    
    dw = (np.dot((a-Y),X.T))/m
    db = np.sum(a-Y)/m
    
    grads={"dw":dw, "db":db}
    return grads, loss
    
```

```python
#test
w, b, X, Y = np.array([[1.],[2.]]), 2., np.array([[1.,2.,-1.],[3.,4.,-3.2]]), np.array([[1,0,1]])
grads, cost = propagate(w, b, X, Y)
print ("dw = " + str(grads["dw"]))
print ("db = " + str(grads["db"]))
print ("cost = " + str(cost))
```

```python
#optimize

def optimize(w,b,X,Y, num_iter, lr, print_cost=False):
    
    costs=[]
    for i in range(num_iter):
        grads, loss= propagate(w,b,X,Y)
        dw = grads['dw']
        db = grads['db']
        
        w = w - (lr * dw).T
        b = b - (lr * db).T
        
        if i %100 ==0:
            costs.append(loss)
#         print(loss)
        if print_cost and i %100 ==0:
            print('Loss after iter %i : %f'%(i, loss ))
            
    params ={'w':w, 'b' :b}
    grads = {'dw':dw, 'db':db}
    
    return params, grads, costs
        
```

```python
w, b, X, Y = np.array([[1.],[2.]]), 2., np.array([[1.,2.,-1.],[3.,4.,-3.2]]), np.array([[1,0,1]])
optimize(w,b,X,Y, num_iter=2000, lr=0.0005,print_cost=True)
```

```python
def predict(w,b,X):
    
    m = X.shape[1]
    Y_predict = np.zeros((1,m))
    w = w.reshape(X.shape[0],1)
    a = sigmoid(np.dot(w.T,X)+b)

    for i in range(m):
        if round(a[0][i])==1:
            Y_predict[0][i] = 1
        else:
            Y_predict[0][i] =0
            
    return Y_predict
```

```python

# model

def logistic_regression(X_train,Y_train, X_test,Y_test=None, num_iter=2000,
                       lr=0.5, print_cost=False):

    w,b = initialize_zeros(X_train.shape[0])
    params, grads, cost = optimize(w,b,X_train, Y_train, num_iter= num_iter,lr=lr, print_cost = print_cost)
    w= params['w']
    b= params['b']
    predict_train = predict(w,b,X_train)
    train_accuracy = 100*(np.mean(1-(np.abs(Y_train-predict_train))))
    predict_test = predict(w,b,X_test)
    if Y_test.all()!=None:
        test_accuracy =  100*(np.mean(1-(np.abs(Y_test-predict_test))))
    
    res = {'cost':cost[-1],'learning_rate':lr, 'num_iter':num_iter,
           'train_acrc':train_accuracy, 'test_acrc':test_accuracy}
    return res
        
```
